```python
#summing columns of same type
df.columns = [i.replace(' [kW]', '') for i in df.columns] #to remove the ' [kW]' suffix from each column
df['Furnace'] = df[['Furnace 1','Furnace 2']].sum(axis=1)
df['Kitchen'] = df[['Kitchen 12','Kitchen 14','Kitchen 38']].sum(axis=1)
df.drop(['Furnace 1','Furnace 2','Kitchen 12','Kitchen 14','Kitchen 38','icon','summary'], axis=1, inplace=True)
```

```python
df = df[0:-1] #to remove the last row
df.tail()
```

```python
df['cloudCover'] = df['cloudCover'].replace('cloudCover', pd.NA) #to replace cloudCover expressions by NaN
df['cloudCover'] = pd.to_numeric(df['cloudCover'], errors='coerce') #to convert the values in the 'cloudCover' column of the DataFrame 'df' to numeric data type
```

```python
# Convert Unix timestamp to readable date
import datetime
timestamp = datetime.datetime.fromtimestamp(1451624400)
print(timestamp)
```

```python
df['time'] = pd.DatetimeIndex(pd.date_range('2016-01-01 05:00', periods=len(df),  freq='min')) #upgrade the time column with a readable date
```

```python
#to extract the different time columns
df['year'] = df['time'].apply(lambda x : x.year)
df['month'] = df['time'].apply(lambda x : x.month)
df['day'] = df['time'].apply(lambda x : x.day)
df['weekday'] = df['time'].apply(lambda x : x.weekday())
df['weekofyear'] = df['time'].apply(lambda x : x.weekofyear)
df['hour'] = df['time'].apply(lambda x : x.hour)
df['minute'] = df['time'].apply(lambda x : x.minute)
```

```python
#to categorize hours into different time periods based on the range
def hours_to_timing(x):
    if x in [22,23,0,1,2,3]:
        timing = 'Night'
    elif x in range(4, 12):
        timing = 'Morning'
    elif x in range(12, 17):
        timing = 'Afternoon'
    elif x in range(17, 22):
        timing = 'Evening'
    else:
        timing = 'X'
    return timing
df['timing'] = df['hour'].apply(hours_to_timing)
```

```python
#to convert timings into integers from 1 to 4
def timing_to_int(timing):
    if timing == 'Night':
        return 1
    elif timing == 'Morning':
        return 2
    elif timing == 'Afternoon':
        return 3
    elif timing == 'Evening':
        return 4
    else:
        return 0

df['timing'] = df['timing'].apply(timing_to_int)
```

```python
#to delete unnecessary columns with same values
df['use_overal'] = df['use']
df['gen_solar'] = df['gen']
df.drop(['use','House overall','gen','Solar'], axis=1, inplace=True)
```

```python
#mean consumption per month
def groupByMonth(column):
    return df[[column,'month']].groupby('month').agg({column:['mean']})[column]

columns_to_treat = [ 'Dishwasher','Kitchen','Home office', 'Fridge', 'Well', 'Microwave', 'Barn', 'Living room', 'Wine cellar', 'Furnace', 'Garage door', 'use_overal', 'gen_solar']

mean_consumption_per_month = pd.DataFrame()

for column in columns_to_treat:
   mean_consumption_per_month[column] = groupByMonth(column)

mean_consumption_per_month
```

```python
#mean consumption per day
def groupByWeekday(col):
    return df[[col,'weekday']].groupby('weekday').agg({col:['mean']})[col]

columns_to_treat = [ 'Dishwasher','Kitchen','Home office', 'Fridge', 'Well', 'Microwave', 'Barn', 'Living room', 'Wine cellar', 'Furnace', 'Garage door', 'use_overal', 'gen_solar']

mean_consumption_per_day = pd.DataFrame()

for col in columns_to_treat:
   mean_consumption_per_day[col] = groupByWeekday(col)
    
mean_consumption_per_day
```

```python
#mean consumption by timing
def groupByTiming(col):
    return df[[col,'timing']].groupby('timing').agg({col:['mean']})[col]

columns_to_treat = [ 'Dishwasher','Kitchen','Home office', 'Fridge', 'Well', 'Microwave', 'Barn', 'Living room', 'Wine cellar', 'Furnace', 'Garage door', 'use_overal', 'gen_solar']

mean_consumption_by_timing = pd.DataFrame()

for col in columns_to_treat:
   mean_consumption_by_timing[col] = groupByTiming(col)
    
mean_consumption_by_timing
```

```python
# Create a list to store curves
curves = []

# Loop over each month
for month in range(1, 13):
    # Filter data for the 10th of the month
    month_data = df[(df['time'].dt.month == month) & (df['time'].dt.day == 10)]
    
    # Group by timing for the 10th of the month data
    month_grouped = month_data.groupby('timing')['use_overal'].sum()
    
    # Create a curve for the month
    curve = hv.Curve(month_grouped, label=f"Month {month}")
    
    # Append curve to list
    curves.append(curve)

# Overlay all curves on a single plot
overlay = hv.Overlay(curves)

# Customize the plot
overlay.opts(
    title="Total Energy Consumption on 10th of Each Month by Timing",
    xlabel="Timing",
    ylabel="Energy Consumption (KW)",
    width=700,
    height=500,
    legend_position='top',
    show_grid=True,
    fontsize={'title': 12, 'labels': 10}
)
```
